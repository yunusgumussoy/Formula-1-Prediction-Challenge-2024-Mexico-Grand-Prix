```python
# lap_data
# Extract days and hours
lap_data['Days'] = lap_data['Time'].apply(lambda x: int(x.split(' ')[0]))
lap_data['Hours'] = lap_data['Time'].apply(lambda x: int(x.split(' ')[2].split(':')[0]))
lap_data['Minutes'] = lap_data['Time'].apply(lambda x: int(x.split(' ')[2].split(':')[1]))

# View the updated DataFrame
print(lap_data)
```

```python
# Extract days and hours
weather_data['Days'] = weather_data['Time'].apply(lambda x: int(x.split(' ')[0]))
weather_data['Hours'] = weather_data['Time'].apply(lambda x: int(x.split(' ')[2].split(':')[0]))
weather_data['Minutes'] = weather_data['Time'].apply(lambda x: int(x.split(' ')[2].split(':')[1]))

# View the updated DataFrame
print(weather_data)
```

```python
# Remove duplicates in weather_data based on 'eventname' and 'hours'
weather_data1 = weather_data.drop_duplicates(subset=['EventName', 'Hours', 'Minutes'])

# Now perform the left join
lap_data1 = pd.merge(lap_data, weather_data1, on=['EventName', 'Hours', 'Minutes'], how='left')

# The number of rows in merged_df should now match df1

print(lap_data1)
```
